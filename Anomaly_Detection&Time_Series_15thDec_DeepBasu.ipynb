# Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples

Answer:

**Anomaly Detection** is a data analysis technique used to identify data points, patterns, or observations that deviate significantly from the expected or normal behavior of a dataset. These unusual observations are known as **anomalies**, **outliers**, or **novelties**. Anomaly detection plays a crucial role in many real-world applications such as fraud detection, network security, fault detection in machines, medical diagnosis, and quality control.

The primary objective of anomaly detection is to distinguish rare and abnormal instances from normal data, especially when anomalies may indicate critical events such as system failures or security breaches.

---

## **Types of Anomalies**

Anomalies can be broadly classified into three main types:

---

## **1. Point Anomalies**

A **point anomaly** occurs when an individual data instance is significantly different from the rest of the data.

### **Explanation**
In this type, a single data point is anomalous when compared to the majority of data points. These anomalies are the simplest and most commonly detected type.

### **Example**
- In a credit card transaction dataset, a single transaction of ₹5,00,000 made from a normally low-spending customer could be considered a point anomaly.
- In a temperature monitoring system, a sudden reading of 90°C when the normal range is 20–30°C is a point anomaly.

### **Use Case**
Point anomalies are widely used in **fraud detection**, **sensor error detection**, and **intrusion detection systems**.

---

## **2. Contextual Anomalies**

A **contextual anomaly** (also known as a conditional anomaly) is a data point that is considered anomalous only within a specific context.

### **Explanation**
In this case, the anomaly depends on contextual attributes such as time, location, or season. A data point may be normal in one context but anomalous in another.

### **Example**
- A temperature of 30°C is normal during summer but would be considered anomalous during winter.
- High network traffic at midnight might be unusual, whereas the same traffic during business hours could be normal.

### **Use Case**
Contextual anomalies are common in **time-series data**, **weather forecasting**, **health monitoring**, and **user behavior analysis**.

---

## **3. Collective Anomalies**

A **collective anomaly** occurs when a group of related data points is anomalous, even if individual data points appear normal on their own.

### **Explanation**
Here, the anomaly is identified by analyzing a sequence or collection of data points rather than individual observations.

### **Example**
- A sequence of small but continuous network packets may indicate a denial-of-service (DoS) attack, even though each packet individually looks normal.
- In ECG data, a pattern of heartbeats may indicate a medical abnormality, even if individual beats are within normal limits.

### **Use Case**
Collective anomalies are important in **network security**, **industrial process monitoring**, and **medical signal analysis**.

---

## **Conclusion**

Anomaly detection is a vital technique for identifying unusual and potentially harmful patterns in data. **Point anomalies** focus on individual deviations, **contextual anomalies** depend on specific conditions, and **collective anomalies** identify abnormal patterns in groups of data points.

# Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.


Answer:

## **Comparison of Isolation Forest, DBSCAN, and Local Outlier Factor (LOF)**

Anomaly detection techniques differ in their underlying approach, assumptions, and suitability for different types of datasets. **Isolation Forest**, **DBSCAN**, and **Local Outlier Factor (LOF)** are three widely used methods, each with unique strengths and application areas. The comparison below explains their working principles, advantages, limitations, and suitable use cases.

---

## **1. Isolation Forest**

### **Approach**
Isolation Forest is a **tree-based ensemble method** specifically designed for anomaly detection. Unlike traditional methods that model normal data, Isolation Forest works by **isolating anomalies**. It randomly selects features and splits values to create decision trees.  
Anomalies are isolated faster because they are fewer and more distinct, requiring fewer splits to separate them from the rest of the data.

### **Key Characteristics**
- Does not rely on distance or density measures  
- Works efficiently with high-dimensional data  
- Requires minimal parameter tuning  
- Scales well to large datasets  

### **Suitable Use Cases**
- Fraud detection in financial transactions  
- Network intrusion detection  
- Large-scale datasets with many features  
- Situations where anomalies are rare and significantly different from normal data  

---

## **2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**

### **Approach**
DBSCAN is a **density-based clustering algorithm** that identifies clusters based on the density of data points. Points that lie in low-density regions are labeled as anomalies or noise.

It relies on two key parameters:
- **Epsilon (ε):** Maximum distance between two points to be considered neighbors  
- **MinPts:** Minimum number of points required to form a dense region  

### **Key Characteristics**
- Does not require specifying the number of clusters in advance  
- Can identify arbitrarily shaped clusters  
- Sensitive to parameter selection  
- Struggles with datasets of varying densities  

### **Suitable Use Cases**
- Spatial data analysis (e.g., geographic clustering)  
- Anomaly detection where anomalies are isolated from dense regions  
- Applications where cluster shapes are irregular  

---

## **3. Local Outlier Factor (LOF)**

### **Approach**
Local Outlier Factor is a **density-based anomaly detection algorithm** that compares the local density of a data point with the densities of its neighbors. A point is considered an outlier if it has significantly lower density than its surrounding points.

### **Key Characteristics**
- Detects local anomalies effectively  
- Works well when anomalies are subtle and relative  
- Sensitive to the choice of neighborhood size (k)  
- Computationally expensive for large datasets  

### **Suitable Use Cases**
- Detecting local deviations in complex datasets  
- Fraud detection where anomalies occur in dense regions  
- Sensor data analysis and monitoring systems  

---

## **Comparative Summary**

| Feature | Isolation Forest | DBSCAN | Local Outlier Factor |
|------|----------------|--------|----------------------|
| Approach | Tree-based isolation | Density-based clustering | Local density comparison |
| Detects Local Anomalies | Moderate | Limited | Excellent |
| Scalability | High | Medium | Low–Medium |
| Handles High Dimensions | Yes | Poor | Moderate |
| Parameter Sensitivity | Low | High | Moderate |
| Best Use Case | Large, high-dimensional datasets | Spatial and density-based data | Contextual/local anomalies |

# Question 3: What are the key components of a Time Series? Explain each with one example.

Answer:

A **time series** is a sequence of data points recorded at regular time intervals. Time series analysis helps in understanding underlying patterns, forecasting future values, and identifying trends or irregularities in data over time. A time series is generally composed of four main components:

---

## **1. Trend**

### **Explanation**
The **trend** represents the long-term movement or direction in a time series. It shows whether the data is generally increasing, decreasing, or remaining constant over a long period.

### **Example**
- The gradual increase in annual global average temperature over several decades.
- A steady rise in e-commerce sales over multiple years.

---

## **2. Seasonality**

### **Explanation**
**Seasonality** refers to regular, repeating patterns that occur at fixed intervals due to seasonal factors such as weather, festivals, or business cycles.

### **Example**
- Higher sales of air conditioners during summer months every year.
- Increased retail sales during festive seasons like Diwali or Christmas.

---

## **3. Cyclical Component**

### **Explanation**
The **cyclical component** represents long-term oscillations in a time series that do not have a fixed or predictable period. These fluctuations are often influenced by economic or business cycles.

### **Example**
- Periods of economic expansion followed by recession in a country’s GDP.
- Fluctuations in employment rates over several years.

---

## **4. Irregular (Random) Component**

### **Explanation**
The **irregular component** consists of unpredictable, random variations caused by unforeseen events such as natural disasters, strikes, or sudden policy changes.

### **Example**
- A sudden drop in airline bookings due to a natural disaster.
- Unexpected spike in demand due to a viral social media trend.

---

## **Conclusion**

A time series is made up of **trend**, **seasonal**, **cyclical**, and **irregular** components. Understanding these components helps analysts accurately model data, make reliable forecasts, and identify unusual patterns in real-world applications such as finance, weather forecasting, and business analytics.

# Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?

Answer:

## **Stationarity in Time Series**

A **time series is said to be stationary** if its statistical properties such as **mean, variance, and autocovariance remain constant over time**. In a stationary series, the behavior of the data does not depend on the time at which it is observed, making it suitable for most time series forecasting models.

---

## **Characteristics of a Stationary Time Series**

A time series is considered stationary if:

1. **Constant Mean** – The average value does not change over time.  
2. **Constant Variance** – The variability of the series remains stable.  
3. **Constant Autocovariance** – The relationship between values depends only on the time lag, not on the actual time.

---

## **Why Stationarity is Important**

Most time series models such as **ARIMA, ARMA, and SARIMA** assume that the data is stationary. Non-stationary data can lead to unreliable predictions and misleading statistical inferences.

---

## **Testing for Stationarity**

### **1. Augmented Dickey-Fuller (ADF) Test**
The ADF test checks the presence of a unit root in the time series.

- **Null Hypothesis (H₀):** The time series is non-stationary  
- **Alternative Hypothesis (H₁):** The time series is stationary  

**Interpretation:**
- If *p-value < 0.05*, reject H₀ → series is stationary  
- If *p-value ≥ 0.05*, fail to reject H₀ → series is non-stationary  

---

### **2. Kwiatkowski–Phillips–Schmidt–Shin (KPSS) Test**
- Tests whether a time series is stationary around a mean or trend.
- Opposite of ADF:
  - **H₀:** Series is stationary  
  - **H₁:** Series is non-stationary  

Using both ADF and KPSS together gives more reliable results.

---

## **Transforming a Non-Stationary Series into a Stationary One**

### **1. Differencing**
Subtracting the previous observation from the current one removes trend and seasonality.

\[
Y'_t = Y_t - Y_{t-1}
\]

Used commonly in ARIMA models.

---

### **2. Logarithmic Transformation**
Applying log transformation stabilizes variance, especially when data grows exponentially.

\[
Y'_t = \log(Y_t)
\]

---

### **3. Detrending**
Removing the trend component using regression or moving averages.

---

### **4. Seasonal Differencing**
Removes seasonal effects by subtracting values from previous seasons.

\[
Y'_t = Y_t - Y_{t-s}
\]

where *s* is the seasonal period.

# Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.

Answer:

## **Difference Between AR, MA, ARIMA, SARIMA, and SARIMAX Models**

Time series forecasting models differ in how they capture patterns such as trends, seasonality, and external influences. The following models are widely used for forecasting temporal data, each with its own structure and application.

---

## **1. AR (AutoRegressive) Model**

### **Structure**
The **AutoRegressive (AR)** model predicts the current value of a time series using its **own past values**.

\[
Y_t = c + \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \dots + \phi_p Y_{t-p} + \epsilon_t
\]

Where:
- \( p \) = number of lagged observations
- \( \epsilon_t \) = white noise

### **Application**
- Suitable when the data shows **strong correlation with its past values**
- Used in short-term forecasting of stationary time series

---

## **2. MA (Moving Average) Model**

### **Structure**
The **Moving Average (MA)** model uses **past forecast errors** instead of past values.

\[
Y_t = \mu + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + \dots + \theta_q \epsilon_{t-q}
\]

Where:
- \( q \) = number of lagged error terms

### **Application**
- Useful when random shocks or noise affect the data
- Effective for smoothing short-term fluctuations

---

## **3. ARIMA (AutoRegressive Integrated Moving Average)**

### **Structure**
ARIMA combines **AR**, **I (Integrated)**, and **MA** components.

\[
ARIMA(p, d, q)
\]

- **p**: number of autoregressive terms  
- **d**: number of differences applied to make the series stationary  
- **q**: number of moving average terms  

### **Application**
- Suitable for **non-stationary time series**
- Widely used in economic and financial forecasting
- Works well when data shows trend but no strong seasonality

---

## **4. SARIMA (Seasonal ARIMA)**

### **Structure**
SARIMA extends ARIMA by incorporating **seasonal components**.

\[
SARIMA(p,d,q)(P,D,Q)_s
\]

Where:
- \( (p,d,q) \) = non-seasonal components  
- \( (P,D,Q) \) = seasonal components  
- \( s \) = length of the seasonal cycle  

### **Application**
- Used for data with **both trend and seasonality**
- Common in monthly or quarterly sales, weather, and demand forecasting

---

## **5. SARIMAX (Seasonal ARIMA with Exogenous Variables)**

### **Structure**
SARIMAX extends SARIMA by including **exogenous (external) variables** that influence the target time series.

\[
Y_t = SARIMA(p,d,q)(P,D,Q)_s + X_t
\]

Where:
- \( X_t \) represents external predictors such as promotions, holidays, or economic indicators

### **Application**
- Used when external factors significantly affect the time series
- Ideal for forecasting with additional explanatory variables

---

## **Comparative Summary Table**

| Model | Trend | Seasonality | External Variables | Use Case |
|------|-------|-------------|--------------------|----------|
| AR | No | No | No | Simple stationary data |
| MA | No | No | No | Noise smoothing |
| ARIMA | Yes | No | No | Non-stationary data |
| SARIMA | Yes | Yes | No | Seasonal time series |
| SARIMAX | Yes | Yes | Yes | Time series with external factors |

# Question 8: Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.


```python
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

uploaded = files.upload()

filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

print("Columns in dataset:", df.columns)

date_col = df.columns[0]
value_col = df.columns[1]

df[date_col] = pd.to_datetime(df[date_col])
df.set_index(date_col, inplace=True)

model = SARIMAX(
    df[value_col],
    order=(1,1,1),
    seasonal_order=(1,1,1,12),
    enforce_stationarity=False,
    enforce_invertibility=False
)

results = model.fit()

forecast = results.get_forecast(steps=12)
forecast_index = pd.date_range(
    start=df.index[-1] + pd.offsets.MonthBegin(),
    periods=12,
    freq='MS'
)

plt.figure(figsize=(10,6))
plt.plot(df.index, df[value_col], label="Actual")
plt.plot(forecast_index, forecast.predicted_mean, label="Forecast", color="red")
plt.fill_between(
    forecast_index,
    forecast.conf_int().iloc[:,0],
    forecast.conf_int().iloc[:,1],
    color="pink",
    alpha=0.3
)
plt.title("SARIMA Forecast – AirPassengers Dataset")
plt.xlabel("Date")
plt.ylabel("Passengers")
plt.legend()
plt.grid(True)
plt.show()
```

# Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.

```python
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

uploaded = files.upload()

filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

print("Columns in dataset:", df.columns)

numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns
df_numeric = df[numeric_cols].dropna()

lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
df_numeric["anomaly"] = lof.fit_predict(df_numeric)

df_numeric["anomaly"] = df_numeric["anomaly"].map({1: 0, -1: 1})

plt.figure(figsize=(10,6))
plt.scatter(
    df_numeric.iloc[:, 0],
    df_numeric.iloc[:, 1],
    c=df_numeric["anomaly"],
    cmap="coolwarm",
    s=50
)
plt.xlabel(df_numeric.columns[0])
plt.ylabel(df_numeric.columns[1])
plt.title("Local Outlier Factor (LOF) Anomaly Detection")
plt.colorbar(label="Anomaly (1 = Outlier)")
plt.grid(True)
plt.show()
```

# Question 10: You are working as a data scientist for a power grid monitoring company. Your goal is to forecast energy demand and also detect abnormal spikes or drops in real-time consumption data collected every 15 minutes. The dataset includes features like timestamp, region, weather conditions, and energy usage.
Explain your real-time data science workflow:

● How would you detect anomalies in this streaming data (Isolation Forest / LOF / DBSCAN)?

● Which time series model would you use for short-term forecasting (ARIMA /
SARIMA / SARIMAX)?

● How would you validate and monitor the performance over time?

● How would this solution help business decisions or operations?


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.arima.model import ARIMA


# Sample energy data

np.random.seed(0)
time = pd.date_range("2023-01-01", periods=200, freq="15min")
energy = 100 + np.random.normal(0, 5, 200)
energy[60:65] += 40   # spike anomaly
energy[130:135] -= 30 # drop anomaly

df = pd.DataFrame({"energy": energy}, index=time)


iso = IsolationForest(contamination=0.05, random_state=42)
df["anomaly"] = iso.fit_predict(df[["energy"]])
df["anomaly"] = df["anomaly"].map({1: 0, -1: 1})


model = ARIMA(df["energy"], order=(1,1,1))
fit = model.fit()
forecast = fit.forecast(steps=12)


plt.figure(figsize=(10,5))
plt.plot(df.index, df["energy"], label="Energy Usage")
plt.scatter(df[df["anomaly"]==1].index,
            df[df["anomaly"]==1]["energy"],
            color="red", label="Anomaly")
plt.plot(pd.date_range(df.index[-1], periods=13, freq="15min")[1:],
         forecast, label="Forecast", color="green")
plt.legend()
plt.title("Energy Consumption: Anomaly Detection & Forecasting")
plt.show()
```

### 1. How would you detect anomalies in this streaming data?
To detect anomalies in real-time energy consumption data, **Isolation Forest** is used. It works by isolating data points that behave differently from the majority. Since abnormal spikes or drops in energy usage are rare events, they get isolated faster in the tree structure, making Isolation Forest well-suited for streaming and high-volume data. It does not require labeled data and performs efficiently in real-time environments.

---

### 2. Which time series model would you use for short-term forecasting?
For short-term forecasting, **ARIMA (AutoRegressive Integrated Moving Average)** is the most suitable model. It captures short-term trends and temporal dependencies effectively and works well with limited historical data. ARIMA is computationally efficient and ideal when seasonality is minimal or short-term predictions are required.

---

### 3. How would you validate and monitor the performance over time?
Model performance can be validated using:
- Error metrics such as **MAE**, **RMSE**, and **MAPE**
- Rolling or sliding window evaluation to track performance over time
- Monitoring residuals and anomaly frequency to detect data drift
- Periodic retraining using recent data to maintain accuracy

---

### 4. How would this solution help business operations?
This approach enables:
- Early detection of abnormal energy usage or equipment faults  
- Reduced downtime through proactive maintenance  
- Better demand forecasting and load balancing  
- Improved operational efficiency and cost optimization  
- Data-driven decision-making for energy planning and resource management
