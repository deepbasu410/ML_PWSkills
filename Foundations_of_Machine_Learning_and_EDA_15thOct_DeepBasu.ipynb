# Question 1 : What is the difference between AI, ML, DL, and Data Science? Provide a brief explanation of each.
(Hint: Compare their scope, techniques, and applications for each.)

Answer:

***Artificial Intelligence (AI)-***

Artificial Intelligence is a broad area of computer science that aims to create systems capable of performing tasks that require human intelligence such as reasoning, problem-solving, and learning.

**Scope:** Includes machine learning, deep learning, natural language processing, and robotics.

**Techniques:** Expert systems, rule-based systems, NLP, computer vision, robotics.

**Applications:** Chatbots (Siri, Alexa), autonomous vehicles, fraud detection, and intelligent assistants.

***Machine Learning (ML)-***

Machine Learning is a subset of AI which enables systems to automatically learn and improve from experience without explicit programming.

**Scope:** Focuses on algorithms that learn from historical data to make predictions.

**Techniques:**

  * Supervised Learning (Linear Regression, Decision Trees)

  * Unsupervised Learning (K-Means, PCA)

  * Reinforcement Learning

**Applications:** Email spam filters, stock prediction, recommendation systems, predictive maintenance.

***Deep Learning (DL)-***

Deep Learning is a subfield of ML that uses multi-layered neural networks to automatically learn complex patterns from large volumes of data.

**Scope:** Works effectively with unstructured data such as images, audio, and text.

**Techniques:** CNNs (for images), RNNs (for sequences), Transformers (for text).

**Applications:** Speech recognition, image classification, self-driving cars, face recognition.

***Data Science-***

Data Science is an interdisciplinary field that uses statistical analysis, computer programming, and domain knowledge to extract insights from data for decision-making.

**Scope:** Covers data collection, cleaning, visualization, analysis, and model deployment.

**Techniques:** Data preprocessing, statistical analysis, visualization, machine learning models.

**Applications:** Business analytics, customer segmentation, market forecasting, healthcare analytics.

| **Applications** | Chatbots, self-driving cars, fraud detection | Recommendations, price prediction, spam filtering | Image recognition, speech translation, face ID | Business analytics, forecasting, healthcare insights |

# Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them?

Hint: Discuss bias-variance tradeoff, cross-validation, and regularization techniques.


Answer:

In Machine Learning, the goal is to build a model that generalizes well to new, unseen data.
However, models can either overfit (learn too much noise) or underfit (learn too little), leading to poor predictive performance.

***Overfitting-***

**Definition:** Overfitting occurs when a model learns not only the underlying patterns in training data but also the noise and random fluctuations.

**Characteristics:**

  * High accuracy on training data but poor accuracy on test data.

  * Model becomes too complex (e.g., too many parameters or deep trees).

*Example:* A decision tree that memorizes every training point will perform poorly on new data.

**Detection:**

  * Large difference between training and validation accuracy.

  * Validation loss increases while training loss decreases.

**Prevention Methods:**

  * **Cross-Validation:** Evaluate model on multiple folds to ensure consistent performance.

  * **Regularization:** Add penalties to large weights (L1/L2 regularization).

  * **Early Stopping:** Stop training when validation error stops improving.

  * **Simplify Model:** Reduce complexity or number of features.

  * **Dropout (for DL):** Randomly drop neurons during training to prevent co-adaptation.

***Underfitting-***

**Definition:** Underfitting occurs when a model is too simple to capture the underlying pattern in the data.

**Characteristics:**

  * Low accuracy on both training and test data.

  * Model fails to learn even basic relationships.

*Example:* Using a linear regression model to fit highly nonlinear data.

**Detection:**

  * Both training and validation losses are high.

  * Adding more data or epochs does not improve performance.

**Prevention Methods:**

  * **Increase Model Complexity:** Use more features or deeper models.

  * **Reduce Regularization Strength:** Allow model to fit data more flexibly.

  * **Feature Engineering:** Add meaningful features to capture complex patterns.

**The bias–variance tradeoff** is a fundamental concept that explains the relationship between model complexity and prediction error.

  * Bias refers to the error introduced by simplifying real-world problems too much. High bias leads to underfitting because the model fails to capture key patterns.

  * Variance refers to the model’s sensitivity to fluctuations in the training data. High variance leads to overfitting because the model learns random noise.

**Cross-validation** is a powerful method to detect overfitting and estimate a model’s performance on unseen data.
The dataset is divided into multiple subsets (or “folds”), and the model is trained on some folds and tested on the remaining ones in rotation.
The most common method is **k-fold cross-validation**, where the data is split into k parts (e.g., 5 or 10), and each part serves as the test set once.

**Regularization** is a set of techniques used to prevent overfitting by adding a penalty to large model parameters.

Common regularization methods include:

1. L1 Regularization (Lasso): Adds the absolute value of coefficients as a penalty. It can reduce some weights to zero, effectively performing feature selection.

2. L2 Regularization (Ridge): Adds the square of coefficients as a penalty.
It prevents coefficients from becoming too large but does not eliminate them.

3. Elastic Net: Combines both L1 and L2 penalties to balance coefficient shrinkage and sparsity.

In neural networks, *dropout* is also a regularization method where random neurons are ignored during training to prevent over-dependence on specific pathways.

# Question 3: How would you handle missing values in a dataset? Explain at least three methods with examples.

Hint: Consider deletion, mean/median imputation, and predictive modeling.


Answer:

Handling missing data correctly is crucial because it can affect the accuracy, reliability, and generalization of a machine learning model. There are several methods to handle missing data depending on its type, amount, and importance. Common approaches include deletion, imputation, and predictive modeling.

***Deletion Methods-***

**Definition:** Deletion involves removing the records or features that contain missing values. It is a simple and commonly used technique when the proportion of missing data is small.

Types:

1. **Listwise Deletion (Complete Case Analysis):** In this approach, any row with one or more missing values is completely removed from the dataset.
This method is suitable when missing data is less than 5% and randomly distributed.

    *Example:* Suppose we have 1,000 customer records, and 20 records have missing income values. We can safely delete those 20 rows if their removal doesn’t bias the dataset.

2. **Pairwise Deletion:** Instead of removing entire rows, pairwise deletion uses all available data for each analysis. For example, while calculating correlation between two features, only the pairs with non-missing values are used.

Limitations: Deletion can reduce the sample size, leading to loss of valuable information or biased results if the data is not missing completely at random.

***Mean/Median/Mode Imputation-***

**Definition:** Imputation replaces missing values with estimated ones instead of deleting data. This method is useful when data loss must be minimized and missing values are not too frequent.

Approaches:

1. **Mean Imputation:** For numerical variables, missing values are replaced with the mean of the observed values.

      *Example:* If the “Age” column has missing entries and the average age is 35, we fill all missing “Age” values with 35.

2. **Median Imputation:** Median is preferred when data is skewed or contains outliers because it represents the central tendency better.

      *Example:* In a dataset with customer incomes where outliers exist (like a few very high earners), missing values can be replaced by the median income instead of the mean.

3. **Mode Imputation:** For categorical variables, missing values are replaced with the most frequent category.

      *Example:* If the column “Gender” has missing values and the most frequent value is “Male,” then missing entries can be filled with “Male.”

Limitations: While simple, these methods can reduce variability and distort statistical relationships because imputed values may not reflect real-world variation.

***Predictive Modeling Imputation-***

**Definition:** Predictive modeling uses machine learning algorithms to predict the missing values based on other available variables.
This approach treats missing data as a prediction problem.

**How It Works:**

1. Separate the dataset into two parts: one where the target feature has values and one where it’s missing.

2. Use the first subset to train a model (e.g., regression, decision tree, or k-nearest neighbors).

3. Use the trained model to predict the missing values in the second subset.

*Example:*

If “Salary” values are missing, we can train a regression model using related features like “Experience,” “Age,” and “Education Level” to predict missing salaries.
Similarly, for categorical data, classification models such as Decision Trees can predict missing categories.

Advantages:

  * Maintains relationships between variables.

  * Can handle both numerical and categorical data effectively.

Limitations:

  * Computationally expensive for large datasets.

  * The accuracy of imputation depends on the quality of the predictive model.

# Question 4:What is an imbalanced dataset? Describe two techniques to handle it(theoretical + practical).

Hint: Discuss SMOTE, Random Under/Oversampling, and class weights in models.


Answer:

An imbalanced dataset occurs when one class (majority) has far more samples than another (minority).
*Example:* 95% “No Disease” vs. 5% “Disease”.
Such imbalance causes models to predict the majority class more often, ignoring important minority cases.

***Theoretical Techniques-***

**1. Sampling Methods-**

**a) Random Undersampling:**

Reduces samples from the majority class to match the minority count.

  * *Advantage:* Simple and reduces training time.

  * *Disadvantage:* May lose important majority information.

**b) Random Oversampling:**

Duplicates minority class samples until classes are balanced.

  * *Advantage:* Retains all data.

  * *Disadvantage:* Can cause overfitting as same samples repeat.

**c) SMOTE (Synthetic Minority Oversampling Technique):**

Generates synthetic minority examples by interpolating between nearby minority samples.

  * *Advantage:* More diverse and realistic samples.

  * *Disadvantage:* May create noisy or borderline samples.

**2. Algorithm-Level Method (Class Weights)-**

Instead of modifying the dataset, we adjust the model by giving higher weights to minority class samples. The model learns that misclassifying minority cases is costlier.

  * *Advantage:* Keeps data intact.

  * *Disadvantage:* Needs tuning to avoid overcompensation.

***Practical Techniques-***

**1. Sampling Implementation (SMOTE + Random Sampling)-**
```python
from imblearn.over_sampling import SMOTE, RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler

# Random Oversampling
ros = RandomOverSampler()
X_res, y_res = ros.fit_resample(X, y)

# Random Undersampling
rus = RandomUnderSampler()
X_res, y_res = rus.fit_resample(X, y)

# SMOTE
sm = SMOTE()
X_res, y_res = sm.fit_resample(X, y)
```
These methods are part of the imbalanced-learn library and are effective preprocessing steps before training.

**2. Class Weights Implementation-**
```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(class_weight='balanced')
model.fit(X_train, y_train)
```

The 'balanced' option automatically adjusts weights based on class frequency, improving recall on minority classes.

# Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.

Hint: Explain impact on distance-based algorithms (e.g., KNN, SVM) and gradient
descent.


Answer:

Feature scaling is the process of normalizing the range of independent variables or features. It ensures all features contribute equally to model training and distance calculations.

**Importance of Feature Scaling-**

  * Many ML algorithms (especially distance- and gradient-based) are sensitive to feature magnitude.

  * Scaling improves:

    * Model convergence speed (in gradient descent)

    * Accuracy (in distance-based models like KNN, SVM)

    * Stability (by preventing numerical overflow or bias)

**1. Min–Max Scaling (Normalization)-**

**Formula:**  
\[
X' = \frac{X - X_{min}}{X_{max} - X_{min}}
\]

**Range:** `[0, 1]`  
**When to Use:** Data with known bounds or in neural networks.  
Limitation: Highly sensitive to outliers.  

*Example:*

| Feature | Before | After (Min–Max) |
|:---------|:--------|:----------------|
| Height (cm) | 150–200 | 0–1 |

**2. Standardization (Z-Score Scaling)-**

**Formula:**  
\[
X' = \frac{X - \mu}{\sigma}
\]
where μ = mean, σ = standard deviation.

**Range:** Mean = 0, Std Dev = 1 (unbounded).  
**When to Use:** When data contains outliers or unknown ranges.  
Advantage: Robust and widely applicable.

**3. Impact on Algorithms-**

**a) Distance-Based Algorithms (KNN, SVM):**

These rely on Euclidean distance or similarity metrics.

  * Without scaling, features with large units (e.g., income vs. age) dominate distance.

  * Scaling ensures fair contribution of each feature in distance computation → more accurate clusters and classifications.

**b) Gradient Descent (e.g., Linear/Logistic Regression, Neural Nets):**

  * Gradient descent updates weights proportional to feature magnitudes.

  * If features are on different scales, convergence becomes slow and unstable.

  * Scaling ensures faster and smoother convergence.

# Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?
Hint: Consider categorical variables with ordinal vs. nominal relationships.


Answer:

Machine Learning algorithms work best with **numerical data**.  
However, categorical features often appear as text (e.g., “Red”, “Blue”, “Green”).  **Encoding** converts these categories into numeric form so models can interpret them correctly. Two common techniques are **Label Encoding** and **One-Hot Encoding**.

***Label Encoding-***

**Definition:**  
Label Encoding assigns each unique category an **integer value**.

*Example:*
| Category | Encoded Value |
|:----------|:--------------|
| Low | 0 |
| Medium | 1 |
| High | 2 |

*Python Example:*
```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['encoded'] = le.fit_transform(data['Category'])
```


***One-Hot Encoding-***

**Definition:**
One-Hot Encoding converts each unique category into **separate binary columns (0 or 1)** indicating the presence of that category.

*Example:*

If a column called *Category* has values **Low**, **Medium**, and **High**, then after One-Hot Encoding, it becomes:

```
Low   Medium   High
1     0        0
0     1        0
0     0        1
```

*Python Example:*
```python
import pandas as pd

# Example dataset
data = pd.DataFrame({'Category': ['Low', 'Medium', 'High']})

# One-Hot Encoding
encoded = pd.get_dummies(data['Category'])
print(encoded)
```

In short, **Label Encoding** is best for ordinal features where order matters. Whereas, **One-Hot Encoding** is best for nominal features with no order.

# Question 7: Google Play Store Dataset
a). Analyze the relationship between app categories and ratings. Which categories have the highest/lowest average ratings, and what could be the possible reasons?

Dataset: https://github.com/MasteriNeuron/datasets.git


````python
from google.colab import files
uploaded = files.upload()
````

````python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load Dataset
df = pd.read_csv('/content/googleplaystore.csv')

# Step 3: Data Cleaning
df = df[['Category', 'Rating']].dropna(subset=['Category', 'Rating'])
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df = df.dropna(subset=['Rating'])

# Step 4: Compute Average Rating by Category
category_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)

# Step 5: Visualization
plt.figure(figsize=(12,6))
category_ratings.plot(kind='bar', color='teal', edgecolor='black')
plt.title('Average App Ratings by Category', fontsize=14, fontweight='bold')
plt.ylabel('Average Rating')
plt.xlabel('App Category')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Step 6: Insights
top_category = category_ratings.idxmax()
bottom_category = category_ratings.idxmin()

print(f" Highest Rated Category: {top_category} ({category_ratings.max():.2f})")
print(f" Lowest Rated Category: {bottom_category} ({category_ratings.min():.2f})")

print("\n Possible Reasons:")
print(f"- {top_category}: May contain high-quality, user-focused apps with useful features and fewer bugs.")
print(f"- {bottom_category}: Could include ad-heavy, unstable, or less engaging apps, leading to lower ratings.")
````

````python
# a) Compare the survival rates based on passenger class (Pclass).

# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load Dataset
df = pd.read_csv('/content/titanic.csv')

# Step 3: Data Cleaning
df = df[['Pclass', 'Survived']].dropna()

# Step 4: Compute Survival Rate per Class
survival_by_class = df.groupby('Pclass')['Survived'].mean().sort_index(ascending=True)

# Step 5: Display Results
print("Survival Rate by Passenger Class:\n")
print(survival_by_class)

# Step 6: Visualization
plt.figure(figsize=(6,4))
survival_by_class.plot(kind='bar', color=['#2ecc71', '#f1c40f', '#e74c3c'], edgecolor='black')
plt.title('Survival Rate by Passenger Class', fontsize=13, fontweight='bold')
plt.xlabel('Passenger Class (1 = Upper, 3 = Lower)')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Step 7: Insights
highest_class = survival_by_class.idxmax()
rate = survival_by_class.max()

print(f"\n Highest Survival Rate: Class {highest_class} ({rate:.2%})")

print("\n Possible Reason:")
print(f" - Passengers in 1st class likely had better access to lifeboats, early evacuation priority,")
print(f" and cabins closer to deck levels, improving survival chances.")
````

# Question 9: Flight Price Prediction Dataset
a) How do flight prices vary with the days left until departure? Identify any exponential price surges and recommend the best booking window.

b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are consistently cheaper/premium, and why?

Dataset: https://github.com/MasteriNeuron/datasets.git

````python
# a) How do flight prices vary with the days left until departure?

# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Step 2: Load Dataset
df = pd.read_csv('/content/flight_price.csv')

# Step 3: Data Cleaning
df = df[['days_left', 'price']].dropna()
df = df[df['days_left'] >= 0]

# Step 4: Analyze Relationship
# Group prices by number of days left
avg_price_by_day = df.groupby('days_left')['price'].mean()

# Step 5: Visualization
plt.figure(figsize=(8,5))
plt.plot(avg_price_by_day.index, avg_price_by_day.values, color='darkorange', linewidth=2)
plt.title('Average Flight Price vs Days Left Until Departure', fontsize=13, fontweight='bold')
plt.xlabel('Days Left Until Departure')
plt.ylabel('Average Price (₹)')
plt.grid(alpha=0.6)
plt.tight_layout()
plt.show()

# Step 6: Detect Exponential Surge
# Check last few days before flight
last_10_days = avg_price_by_day.tail(10)
surge_detected = last_10_days.pct_change().mean() > 0.05  # >5% avg increase per day
if surge_detected:
    print(" Exponential price surge detected in last few days before departure!")

# Step 7: Identify Best Booking Window
min_price_day = avg_price_by_day.idxmin()
print(f"\n Best booking window: Around {min_price_day} days before departure — lowest average fare observed.")

print("\n Insight:")
print(" - Prices generally rise sharply as the departure date approaches (especially within 7 days).")
print(" - Booking at least 20–30 days in advance usually secures better deals, avoiding last-minute surges.")
````

````python
# b) Compare prices across airlines for the same route (e.g., Delhi-Mumbai).

# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load Dataset
df = pd.read_csv('/content/flight_price.csv')

# Step 3: Filter Specific Route (Delhi → Mumbai)
route_df = df[(df['source_city'] == 'Delhi') & (df['destination_city'] == 'Mumbai')]

# Step 4: Data Cleaning
route_df = route_df[['airline', 'price']].dropna()

# Step 5: Compute Average Price per Airline
avg_price_by_airline = route_df.groupby('airline')['price'].mean().sort_values()

# Step 6: Visualization
plt.figure(figsize=(10,5))
avg_price_by_airline.plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('Average Delhi–Mumbai Flight Prices by Airline', fontsize=13, fontweight='bold')
plt.xlabel('Airline')
plt.ylabel('Average Price (₹)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Step 7: Insights
cheapest = avg_price_by_airline.idxmin()
premium = avg_price_by_airline.idxmax()

print(f" Cheapest Airline: {cheapest} (Avg ₹{avg_price_by_airline.min():.0f})")
print(f" Premium Airline: {premium} (Avg ₹{avg_price_by_airline.max():.0f})")

print("\n Possible Reasons:")
print(f" - {cheapest}: Budget carrier offering fewer frills and add-on services.")
print(f" - {premium}: Full-service airline with better in-flight experience, luggage allowance, and flexible tickets.")
````

# Question 10: HR Analytics Dataset
a). What factors most strongly correlate with employee attrition? Use visualizations to show key drivers (e.g., satisfaction, overtime, salary).

b). Are employees with more projects more likely to leave?

Dataset: hr_analytics

````python
# a) What factors most strongly correlate with employee attrition?

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('/content/hr_analytics.csv')

# Ensure target column exists and rename if needed
if 'left' in df.columns:
    df = df.rename(columns={'left': 'Attrition'})
elif 'Attrition' not in df.columns:
    raise KeyError("No 'left' or 'Attrition' column found. Check df.columns")

# Normalize / encode salary if present and non-numeric
if 'salary' in df.columns:
    if not pd.api.types.is_numeric_dtype(df['salary']):
        mapping = {'low': 0, 'medium': 1, 'high': 2, 'Low':0, 'Medium':1, 'High':2}
        df['salary_encoded'] = df['salary'].map(mapping)

        if df['salary_encoded'].isna().any():
            dummies = pd.get_dummies(df['salary'], prefix='salary')

            df['salary_encoded'] = (dummies.mul(range(1, dummies.shape[1]+1), axis=1)).sum(axis=1)
    else:
        df['salary_encoded'] = df['salary']
else:

    df['salary_encoded'] = pd.NA

df['Attrition'] = df['Attrition'].apply(lambda x: 1 if str(x).lower() in ['yes','1','true','left'] else 0)

candidate_features = [
    'satisfaction_level', 'last_evaluation', 'number_project',
    'average_montly_hours', 'time_spend_company', 'Work_accident',
    'promotion_last_5years', 'salary_encoded', 'Attrition'
]
features_present = [f for f in candidate_features if f in df.columns]

corr = df[features_present].corr()

# Plot heatmap (single diagram)
plt.figure(figsize=(9,7))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', center=0, linewidths=0.5)
plt.title('Correlation Heatmap — Key Factors and Attrition', fontsize=13)
plt.tight_layout()
plt.show()
````

````python
# b) Are employees with more projects more likely to leave?

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/hr_analytics.csv')

df = df.rename(columns={'left': 'Attrition'})

# Convert Attrition to numeric (0 = stayed, 1 = left)
df['Attrition'] = df['Attrition'].apply(lambda x: 1 if str(x).lower() in ['yes','1','true','left'] else 0)

# Compute average attrition rate by number of projects
attrition_by_projects = df.groupby('number_project')['Attrition'].mean().reset_index()

# Plot
plt.figure(figsize=(7,5))
sns.barplot(
    x='number_project',
    y='Attrition',
    data=attrition_by_projects,
    hue='number_project',
    palette='coolwarm',
    legend=False,
    edgecolor='black'
)

plt.title('Attrition Rate by Number of Projects', fontsize=13)
plt.xlabel('Number of Projects')
plt.ylabel('Average Attrition Rate')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
````
